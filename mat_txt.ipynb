```python
# 下面是正式开始批量处理啦
# 读取存放批量.mat的文件夹中的所有.mat文件的文件名
mat_filename_list1 = os.listdir(r"D:\jupyter_psychology\Shen268\ROI_FC\session1")
mat_filename_list2 = os.listdir(r"D:\jupyter_psychology\Shen268\ROI_FC\session2")
mat_filename_list3 = os.listdir(r"D:\jupyter_psychology\Shen268\ROI_FC\session3")
mat_filename_list1
```

```python
# session1批量转化 mat 文件为 txt 文件并保存在桌面的 txt_file 文件夹中
for mat_filename in mat_filename_list1:
    matdata = sio.loadmat(r"D:\jupyter_psychology\Shen268\ROI_FC\session1\\" + mat_filename)
    data = matdata["r"]
    row, col = np.diag_indices_from(data)
    data[row,col] = 0
    txt_filename = mat_filename[:5] + ".txt"  # 例如将“0001.mat” 改为 “0001.txt”
    savetxt_path = r"D:\jupyter_psychology\Shen268\ROI_FC_r\session1\\" + txt_filename
    # print(savetxt_path)
    np.savetxt(savetxt_path, data)   # fmt为保存格式，可不填，会以默认格式保存
# session2批量转化 mat 文件为 txt 文件并保存在桌面的 txt_file 文件夹中
for mat_filename in mat_filename_list2:
    matdata = sio.loadmat(r"D:\jupyter_psychology\Shen268\ROI_FC\session2\\" + mat_filename)
    data = matdata["r"]
    row, col = np.diag_indices_from(data)
    data[row,col] = 0
    txt_filename = mat_filename[:5] + ".txt"  # 例如将“0001.mat” 改为 “0001.txt”
    savetxt_path = r"D:\jupyter_psychology\Shen268\ROI_FC_r\session2\\" + txt_filename
    np.savetxt(savetxt_path, data)   # fmt为保存格式，可不填，会以默认格式保存
# session3批量转化 mat 文件为 txt 文件并保存在桌面的 txt_file 文件夹中
for mat_filename in mat_filename_list3:
    matdata = sio.loadmat(r"D:\jupyter_psychology\Shen268\ROI_FC\session3\\" + mat_filename)
    data = matdata["r"]
    row, col = np.diag_indices_from(data)
    data[row,col] = 0
    txt_filename = mat_filename[:5] + ".txt"  # 例如将“0001.mat” 改为 “0001.txt”
    savetxt_path = r"D:\jupyter_psychology\Shen268\ROI_FC_r\session3\\" + txt_filename
    np.savetxt(savetxt_path, data)   # fmt为保存格式，可不填，会以默认格式保存
```
